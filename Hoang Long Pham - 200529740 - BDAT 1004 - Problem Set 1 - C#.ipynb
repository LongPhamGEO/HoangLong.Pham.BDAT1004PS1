<h4>Question 4 C#</h4><p></p>
Write a program in C# Sharp to separate odd and even integers in separate arrays.<p></p>
Go to the editor<p></p>
Test Data :<p></p>
Input the number of elements to be stored in the array :5<p></p>
Input 5 elements in the array :<p></p>
element - 0 : 25<p></p>
element - 1 : 47<p></p>
element - 2 : 42<p></p>
element - 3 : 56<p></p>
element - 4 : 32<p></p>
Expected Output:<p></p>
The Even elements are:<p></p>
42 56 32<p></p>
The Odd elements are :<p></p>
25 47<p></p>

```
//Question 4 C#

int[] element = new int[] { 25, 47, 42, 56, 32 };
var odd = new List<int>();
var even = new List<int>();

foreach (int i in element)
{
    if (i % 2 == 1)
    {
        odd.Add(i);
    }
    else
    {
        even.Add(i);
    }
}

Console.WriteLine("The Even elements are:");
Console.WriteLine("{0}", string.Join(" ", even));  //Connect all elements of even using " "
Console.WriteLine("The Odd elements are:");
Console.WriteLine("{0}", string.Join(" ", odd)); //Connect all elements of odd using " "
```

```
//Question 5 C#
//Part A:
class Functions
{
    public bool Inside(double x, double y, double x1, double y1, double x2, double y2)
    {
        if ( x > x1 && y > y1 && x < x2 && y < y2)
        {
            return true;
        }
        else
        {
            return false;
        }
    }
}

Functions de = new Functions();
bool result = de.Inside(1,1,0,0,2,3);
Console.WriteLine(result);
```

```
//Question 5 C#
//Part B:

double[] point = {1, 1};
double[] first_rectangle = {0.3, 0.5, 1.1, 0.7};
double[] second_rectangle = {0.5, 0.2, 1.1, 2};

bool first_result = de.Inside(point[0], point[1], first_rectangle[0], first_rectangle[1], first_rectangle[2], first_rectangle[3]);
bool second_result = de.Inside(point[0], point[1], second_rectangle[0], second_rectangle[1], second_rectangle[2], second_rectangle[3]);


if (first_result && second_result) // if both first_result and second_result return true value
{
    Console.WriteLine("The point lies in both of the rectangles.");
}
else if (first_result) // if only first_result returns true value
{
    Console.WriteLine("The point lies only in the first rectangle.");
}
else if (second_result) // if only second_result returns true value
{
    Console.WriteLine("The point lies only in the second rectangle.");    
}
else // if both first_result and second_result return false value
{
    Console.WriteLine("The point lies outside both of the rectangles.");
}
```
